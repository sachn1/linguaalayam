# MCP Server — Direct Python Testing

Tests the three MCP tool functions (`exact_lookup`, `fuzzy_lookup`, `semantic_lookup`) directly, bypassing the MCP protocol layer entirely.

Requires the local Postgres container to be running:
```bash
docker start linguaalayam-pg
```

````python
import sys
from pathlib import Path

# ensure project root is on the path when running from notebooks/
sys.path.insert(0, str(Path.cwd().parent))

from dotenv import load_dotenv
load_dotenv(Path.cwd().parent / ".env")
````

````python
# Initialise tools — loads the embedding model (~30s on first run)
from linguaalayam.mcp import server

server._ensure_docker_db()
server._tools = server._init_tools()
print("Tools ready.")
````

````python
# Exact lookup — known headword
print(server.exact_lookup("run"))
````

````python
# Exact lookup — Malayalam headword
print(server.exact_lookup("ഓടുക"))
````

````python
# Fuzzy lookup — typo / near-match
print(server.fuzzy_lookup("runing", threshold=0.2))
````

````python
# Semantic lookup — meaning-based query
print(server.semantic_lookup("to move quickly on foot", top_k=5))
````

````python
# Semantic lookup — corpus-filtered
print(server.semantic_lookup("complete absence of sound", top_k=3, source="olam_enml"))
````

````python
# No-result cases
print(server.exact_lookup("xyzzy"))
print(server.fuzzy_lookup("xyzzy", threshold=0.8))
````
