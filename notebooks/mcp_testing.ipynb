# MCP Server — Direct Python Testing

Two server variants, two sections:

| Section | Server | Requires |
|---|---|---|
| **Part A** | `mcp-server-http` | Nothing — calls `linguaalayam.org` |
| **Part B** | `mcp-server` (local DB) | Local Postgres container running |

Both sections call the tool functions directly, bypassing the MCP protocol layer entirely.

## Part A — HTTP server (hosted, no local DB needed)

````python
import sys
from pathlib import Path

sys.path.insert(0, str(Path.cwd().parent))

from linguaalayam.mcp import server_http

# Optional: override base URL for a self-hosted instance
# import os; os.environ["LINGUAALAYAM_URL"] = "http://localhost:8000"
````

````python
print(server_http.exact_lookup("run"))
````

````python
print(server_http.fuzzy_lookup("runing", threshold=0.2))
````

````python
print(server_http.semantic_lookup("to move quickly on foot", top_k=5))
````

## Part B — Local DB server (contributors only)

Requires the local Postgres container:
```bash
docker start linguaalayam-pg
```

Loads the embedding model (~30s on first run).

````python
from dotenv import load_dotenv
load_dotenv(Path.cwd().parent / ".env")

from linguaalayam.mcp import server

server._ensure_docker_db()
server._tools = server._init_tools()
print("Tools ready.")
````

````python
print(server.exact_lookup("run"))
````

````python
# Malayalam headword
print(server.exact_lookup("ഓടുക"))
````

````python
print(server.fuzzy_lookup("runing", threshold=0.2))
````

````python
print(server.semantic_lookup("to move quickly on foot", top_k=5))
````

````python
# Corpus-filtered
print(server.semantic_lookup("complete absence of sound", top_k=3, source="olam_enml"))
````

````python
# No-result cases
print(server.exact_lookup("xyzzy"))
print(server.fuzzy_lookup("xyzzy", threshold=0.8))
````
